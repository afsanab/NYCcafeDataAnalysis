## Load Yelp Data - Top 5 Rows

Extract and display the first 5 rows from the Yelp JSON dataset.

```python
import json
import pandas as pd
import tarfile
import os

# Path to the tar file
tar_path = r"c:\Users\Afsana\Documents\TheDataSchoolApplication\Yelp-JSON\Yelp JSON\yelp_dataset.tar"
extract_path = r"c:\Users\Afsana\Documents\TheDataSchoolApplication\Yelp-JSON\extracted"

# Extract tar file if not already extracted
if not os.path.exists(extract_path):
    with tarfile.open(tar_path, "r") as tar:
        tar.extractall(extract_path)
    print("✓ Tar file extracted")

# Find and read JSON files
json_files = []
for root, dirs, files in os.walk(extract_path):
    for file in files:
        if file.endswith('.json'):
            json_files.append(os.path.join(root, file))

print(f"✓ Found {len(json_files)} JSON file(s)")

# Read and display top 5 rows from a JSON file
if json_files:
    json_file = json_files[0]  # Read the first JSON file
    print(f"\nReading from: {json_file}\n")
    
    # Load JSON data (assuming JSONL format - one JSON object per line)
    df = pd.read_json(json_file, lines=True)
    print(df.head(5))
```

## Filter for NYC Cafes

Extract and display only the cafes located in New York City from the Yelp dataset.

```python
import json
import pandas as pd
import tarfile
import os

# Path to the tar file
tar_path = r"c:\Users\Afsana\Documents\TheDataSchoolApplication\Yelp-JSON\Yelp JSON\yelp_dataset.tar"
extract_path = r"c:\Users\Afsana\Documents\TheDataSchoolApplication\Yelp-JSON\extracted"

# Extract tar file if not already extracted
if not os.path.exists(extract_path):
    with tarfile.open(tar_path, "r") as tar:
        tar.extractall(extract_path)
    print("✓ Tar file extracted")

# Find and read the business JSON file
business_file = None
for root, dirs, files in os.walk(extract_path):
    for file in files:
        if file.endswith('.json') and 'business' in file.lower():
            business_file = os.path.join(root, file)
            break

if business_file:
    print(f"✓ Reading from: {business_file}\n")

    # Load JSON data (JSONL format - one JSON object per line)
    df = pd.read_json(business_file, lines=True)

    # Filter for NYC cafes
    # Check if city is NYC and categories contain "cafe" or "coffee"
    nyc_mask = df['city'].str.contains('New York', case=False, na=False)
    cafe_mask = df['categories'].fillna('').str.contains('Cafe|Coffee', case=False, regex=True)

    nyc_cafes = df[nyc_mask & cafe_mask]

    print(f"Total businesses in dataset: {len(df)}")
    print(f"NYC cafes found: {len(nyc_cafes)}\n")

    # Display top 5 NYC cafes
    print("Top 5 NYC Cafes:")
    print(nyc_cafes[['name', 'city', 'state', 'categories', 'stars']].head(5))

    # Save to CSV for easier analysis
    output_path = r"c:\Users\Afsana\Documents\TheDataSchoolApplication\yelp_output.csv"
    nyc_cafes.to_csv(output_path, index=False)
    print(f"\n✓ Saved {len(nyc_cafes)} NYC cafes to: {output_path}")
else:
    print("Business JSON file not found in the extracted data.")
```
